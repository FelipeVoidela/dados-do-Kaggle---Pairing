```python
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
mvs = pd.read_csv("Top_Highest_Openings.csv")
```

```python
#1) Quais sÃ£o os 10 maiores distribuidores

contagem_distribuidores = mvs['Distributor'].value_counts()

MaisRepetido = contagem_distribuidores.head(10)

fig, ax = plt.subplots(figsize=(10, 6))

MaisRepetido.plot(kind='barh', ax=ax)

ax.set_xlabel('Distribuidor')
ax.set_ylabel('Contagem')


ax.set_title('Contagem de Distribuidores Mais Frequentes')

plt.show()
```

```python
#2) Qual filme mais lucrou ao longo de sua exibiÃ§Ã£o

mvs = mvs.sort_values(by='Total Gross', ascending=False)

top_10_filmes = mvs.head(10)

fig, ax = plt.subplots(figsize=(10, 5))

ax.barh(top_10_filmes['Release'], top_10_filmes['Total Gross'], color='orange')

ax.set_xlabel('Ganhos')
ax.set_ylabel('Filmes')

ax.set_title('Filmes que mais lucraram')

plt.show()
```

```python
# 3) Quais 10 filmes menos passaram no cinema?

mvs = mvs.sort_values(by='Theaters', ascending=False)

top_10_filmes = mvs.tail(10)

fig, ax = plt.subplots(figsize=(10, 5))

ax.barh(top_10_filmes['Release'], top_10_filmes['Theaters'], color='red')

ax.set_xlabel('Cinemas')
ax.set_ylabel('Filmes')

ax.set_title('10 filmes menos passaram no cinema')

plt.show()
```

```python
#4) Qual a produtora fez mais filmes

contagem_por_produtora = mvs['Distributor'].value_counts()

produtoraMaisRepetida = contagem_por_produtora.head(5)

plt.figure(figsize=(10, 6))
produtoraMaisRepetida.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('DistribuiÃ§Ã£o de Filmes por Produtora')
plt.ylabel('')
plt.show()
```

```python
#5) Como a mÃ©dia do ganho bruto variou ao longo dos anos para os principais distribuidores?

top_distributors =mvs['Distributor'].value_counts().nlargest(5).index
mvs_top_distributors = mvs[mvs['Distributor'].isin(top_distributors)]
plt.figure(figsize=(12, 8))
for distributor, data in mvs_top_distributors.groupby('Distributor'):
    data.groupby(mvs['Date'].dt.year)['Total Gross'].mean().plot(label=distributor)
plt.title('VariaÃ§Ãµes de ganhos totais por Ano dos Principais Distribuidores')
plt.xlabel('Ano')
plt.ylabel('MÃ©dia de Total Gross')
plt.legend()
plt.show()
```

```python
# 6) Como a mÃ©dia de arrecadaÃ§Ã£o por cinema se comporta ao longo do tempo?

plt.plot(mvs['Date'], mvs['Average'], color='red')
plt.title('MÃ©dia de ArrecadaÃ§Ã£o por Cinema ao Longo do Tempo')
plt.xlabel('Data de LanÃ§amento')
plt.ylabel('ArrecadaÃ§Ã£o MÃ©dia por Cinema')
plt.xticks(rotation=45)
plt.show()
```

```python
#7) Como a arrecadaÃ§Ã£o total se distribui ao longo do ano?

mvs['Year'] = pd.to_datetime(mvs['Date']).dt.year

yearly_total_gross = mvs.groupby('Year')['Total Gross'].sum()

plt.bar(yearly_total_gross.index, yearly_total_gross.values, color='gray')
plt.title('ArrecadaÃ§Ã£o Total por Ano')
plt.xlabel('Ano')
plt.ylabel('ArrecadaÃ§Ã£o Total')
plt.show()
```
